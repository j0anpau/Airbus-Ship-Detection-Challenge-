## Multiple Displaying

```python
# these two variables are "the parameters" of this cell
colums = 6
rows = 6

# this function uses the open, resize and array functions we have seen before
load_img = lambda filename: np.array(PIL.Image.open(f"/kaggle/input/airbus-ship-detection/train_v2/{filename}").resize((500, 500)))

_, axes_list = plt.subplots(colums, rows, figsize=(2*rows, 2*colums)) 

for axes in axes_list:
    for ax in axes:
        ax.axis('off')
        img = np.random.choice(train) # take a random filename 
        ax.imshow(load_img(img)) 
        ax.set_title(img)
```
